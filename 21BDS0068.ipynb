```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('//content/21BDS0068Hitters.csv')
```

```python
print("Missing values:")
print(df.isnull().sum())
```

```python
threshold = len(df) * 0.90
df = df.dropna(axis=1, thresh=threshold)
print("Dataset dimensions after removing columns with >10% missing values:", df.shape)
```

```python
df.head(5)
```

```python
print("Dataset dimensions:", df.shape)
```

```python
df = df.fillna(df.ffill())
```

```python
df.describe
```

```python
irrelevant_columns = ['rownames', 'PID']
df = df.drop(irrelevant_columns, axis=1, errors='ignore')
print("Dataset dimensions after dropping irrelevant columns:", df.shape)
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = pd.Categorical(df[col]).codes
```

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns[:3]
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(numerical_cols):
    sns.histplot(df[col], ax=axes[i])
    axes[i].set_title(f'Distribution of {col} - 21BDS0068')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Loop through numeric columns and create boxplots
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

for i, col in enumerate(numeric_columns, 1):
    plt.subplot(4, 5, i)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df[numerical_cols[0]], y=df[numerical_cols[1]])
plt.title(f'Scatter plot: {numerical_cols[0]} vs {numerical_cols[1]} - 21BDS0068')
plt.xlabel(numerical_cols[0])
plt.ylabel(numerical_cols[1])
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Division"], y=df["Hits"])
plt.title("Hits Distribution by Division")
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()  # Categorical variables
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()  # Numerical variables

print("Categorical columns: ", categorical_cols)
print("Numerical columns: ", numerical_cols)
```

```python
sns.pairplot(df[numerical_cols[:3]])  # Select only the first 3 numerical columns
plt.suptitle('Pair Plot of First 3 Numerical Variables - 21BDS0068', y=1.02)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df[numerical_cols[0]], df[numerical_cols[1]], df[numerical_cols[2]])
ax.set_xlabel(numerical_cols[0])
ax.set_ylabel(numerical_cols[1])
ax.set_zlabel(numerical_cols[2])
plt.title('3D Scatter Plot - 21BDS0068')
plt.show()
```

```python
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap - 21BDS0068')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```
